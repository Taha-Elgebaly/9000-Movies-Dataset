```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mymoviedb.csv',lineterminator='\n')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df_sorted = df.sort_values(by='Popularity', ascending=False)

top_20_movies = df_sorted.head(20)

movie_titles = top_20_movies['Title']
popularity_rates = top_20_movies['Popularity']

plt.figure(figsize=(12, 6))
plt.bar(movie_titles, popularity_rates)
plt.xlabel('Movie')
plt.ylabel('Popularity Rate')
plt.title('Top 20 Movies by Popularity')
plt.xticks(rotation=90)

plt.show()
```

## What is the movie with the highest Popularity rate? And what kind is it?

```python
highest_popularity_movie = df[df['Popularity'] == df['Popularity'].max()]
```

```python
movie_title = highest_popularity_movie['Title']
popularity_rate = highest_popularity_movie['Popularity']
genre = highest_popularity_movie['Genre']
```

```python
plt.bar(movie_title, popularity_rate)
plt.xlabel('Movie')
plt.ylabel('Popularity Rate')
plt.title('Movie with the Highest Popularity Rate')

plt.show()
```

## What year was the largest number of films produced?

```python
df['Year'] = pd.to_datetime(df['Release_Date']).dt.year
```

```python
num_of_films_every_year = df['Year'].value_counts().sort_index()
```

```python
plt.plot(num_of_films_every_year.index, num_of_films_every_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Films')
plt.title('Number of Films Produced per Year')

plt.show()


```

```python
# Find the year with the largest number of films produced
year_with_most_films = num_of_films_every_year.idxmax()
year_with_most_films
```
